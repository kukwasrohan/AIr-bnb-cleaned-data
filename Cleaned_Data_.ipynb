# **Missing Data Handling**

```python
# Data Cleaning and Missing Data Handling
# Checking for missing values
missing_values = df.isnull().sum()
missing_values
```

On check found that the missing values contain in the columns of last_review,review_per_month,name,host_name.so lets remove and fill the missing values based on review found that the missing values is less than 90% of the dataset so we need to fill the missing values

```python
# filling the missing values in the last review column
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# Imputing missing values in 'reviews_per_month' column with the mean
mean_reviews_per_month = df['reviews_per_month'].mean()
df['reviews_per_month'].fillna(mean_reviews_per_month, inplace=True)
```

```python
# Imputing missing values in 'last_review' column with the most common date (mode)
most_common_last_review = df['last_review'].mode()[0]
df['last_review'].fillna(most_common_last_review, inplace=True)
```

```python
# Imputing missing values in 'name' column with the Not Aveliable

df['name'].fillna('Not Available',inplace=True)
```

```python
df['host_name'].fillna('Not Available',inplace=True)
```

```python
# Checking for missing values after imputation
missing_values_ = df.isnull().sum()
print(f'the missig values after the imputation {missing_values_}')
```
